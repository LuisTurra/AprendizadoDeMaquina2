```python
atributos = ['sepala_comprimento','sepala_largura','petala_comprimento','petala_largura','variedade']
iris = pd.read_csv('iris.csv',header = None,names = atributos)
```

```python
iris
```

```python
feature_cols= ['sepala_comprimento','sepala_largura','petala_comprimento','petala_largura']
x = iris[feature_cols]
y = iris.variedade
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print('Accuracy: ',metrics.accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
cn=['setosa', 'versicolor', 'virginica']
from sklearn import tree
tree.plot_tree(clf,
           feature_names = feature_cols, 
           class_names=cn,
           filled = True);
```

```python
clf1 = DecisionTreeClassifier(criterion = 'entropy',splitter='random',max_depth=3)
clf1 = clf1.fit(X_train,y_train)
y_pred1 = clf1.predict(X_test)
print('Accuracy: ',metrics.accuracy_score(y_test,y_pred1))
```

```python
clf2 = DecisionTreeClassifier(criterion = 'gini',splitter='best',max_depth=5)
clf2 = clf2.fit(X_train,y_train)
y_pred2 = clf2.predict(X_test)
print('Accuracy: ',metrics.accuracy_score(y_test,y_pred2))
```
